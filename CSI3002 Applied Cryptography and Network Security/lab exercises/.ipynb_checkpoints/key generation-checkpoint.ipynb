```python
def make_readable(x):
    binary_string = ""
    for i in range(0, len(x), 8):
        binary_string = binary_string + " " + "".join(x[i:i + 8])
    return binary_string


def expand(table, text):
    '''
    table = expansion table
    text = binary list
    '''
    result = []
    for i in table:
        result.append(text[i - 1])  # use lc
    return result


def xor(text1, text2):
    result = []
    for i, j in zip(text1, text2):
        if i == j:
            result.append("0")
        else:
            result.append("1")
    return result


def substitute(table, bits):
    sResult = []
    for i, j in zip(table, range(0, 48, 6)):
        chunk = bits[j:j + 6]
        row = int((chunk[0] + chunk[-1]), 2)
        column = bcd2dec("".join(chunk[1:5]))
        # Converts decimal value into
        value = i[row][column]
        sResult.append("{0:b}".format(value).zfill(4))

    return ''.join(sResult)


def permutate(table, sBits):
    pResult = []
    for i in table:
        pResult.append(sBits(i - 1))
    return "".join(pResult)

#########################################################################

def pc1(table, key):
    parityDropped = []
    for i in table:
        parityDropped.append(key[i - 1])
    return parityDropped


def shift(pc1text, round):
    text = pc1text
    for _ in range(2):
        flow1 = text[0]
        flow2 = text[28]
        left = text[1:28]
        right = text[29:]
        left.append(flow1)
        right.append(flow2)
        left.extend(right)
        text = left
        if round in (1, 2, 9, 16):  ## 2 
            break
    return text


def pc2(table, shifted_text):
    pc2_result = []
    for i in table:
        pc2_result.append(shifted_text[i - 1])
    return pc2_result


def subkeys(keytext):
    permutatedc1 = keytext
    for i in range(1, 17):
        permutatedc1 = shift(permutatedc1, i)
        print(f"Left  key C{i}: ",{make_readable(permutatedc1[:28])})
        print(f"Right key D{i}: ",{make_readable(permutatedc1[28:])})
        print("\n")

        permutatedc2 = pc2(pc2, permutatedc1)
        print(f"Subkey K[{i}]: {make_readable(permutatedc2)} \n")


# Driver code
# Plain text = 64bits
pt = "0000000100100011010001010110011110001001101010111100110111101111"
key = "0001001100110100010101110111100110011011101111001101111111110001"
left = pt[:32]  # left part (Li)
right = pt[32:]  # right part (Ri)

permutatedc1 = pc1(pc1, key)
print("Actual Key    : ", make_readable(key))
print("Permutated Key: ", make_readable(permutatedc1))

print("Left  key C0: {}".format(make_readable(permutatedc1[:28])))
print("Right key D0: {}".format(make_readable(permutatedc1[28:])))

subkeys(permutatedc1)

```

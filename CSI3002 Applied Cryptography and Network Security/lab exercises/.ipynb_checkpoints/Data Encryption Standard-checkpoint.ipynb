## 64-bit Plain Text

```python
def BCD(n):
    binary_num = bin(n).replace("0b", "")
    return binary_num
```

```python
## small letters --> 97 to 122
## capital letters --> 65 to 90

def sixty_four_bit_convert(plain_text):
    list1 = []

    ## decimal to binary conversion
    for i in range(len(plain_text)):
        char1 = plain_text[i]
        ascii_value = ord(char1) ## character to ascii
        
        ascii_str    = str(ascii_value)
        first_digit  = int(ascii_str[0]) ## 1st digit --> binary
        second_digit = int(ascii_str[1]) ## 2nd digit --> binary

        list1.append(BCD(first_digit))
        list1.append(BCD(second_digit))
        
    ## binary in-terms of 4 bits
    for i in range(len(list1)):
        element = str(list1[i])
        if len(element)<4:
            diff = 4 - len(element)

            for j in range(diff):
                element = "0" + element

            list1[i] = element

    ## list1 -> 64 bits as separate-separate ([0100,1001])
    return list1
#    return "".join(list1) ## join the 64 bits together (01001001)
```

```python
sixty_four_bit_plaintext = sixty_four_bit_convert(plain_text)
print(sixty_four_bit_plaintext)
print(len(sixty_four_bit_plaintext))
```

### 64bit plain text

```python
def initial_permutation(elements):
    
    ## input  --> 64bits
    ## output --> 64bits
    
    ## initial_perm_matrix --> 1 to 64 bits
    ## 64bit plain text    --> 0 to 63 bits
    
    str_permutation_matrix = [58, 50, 42, 34, 26, 18, 10, 2,
                          60, 52, 44, 36, 28, 20, 12, 4,
                          62, 54, 46, 38, 30, 22, 14, 6,
                          64, 56, 48, 40, 32, 24, 16, 8,
                          57, 49, 41, 33, 25, 17, 9, 1,
                          59, 51, 43, 35, 27, 19, 11, 3,
                          61, 53, 45, 37, 29, 21, 13, 5,
                          63, 55, 47, 39, 31, 23, 15, 7]

    permuted_matrix = [0 for i in range(64)]
    for i in range(0, len(str_permutation_matrix)):
        index              = str_permutation_matrix[i] - 1   ## so subtracting 1
        permuted_matrix[i] = elements[index]
        
    return permuted_matrix
```

```python
def permuted_choice_1(elements): 
    
    ## input -->  64 bits
    ## output --> 56 bits
    
    ## initial_perm_matrix --> 1 to 64 bits
    ## 64bit key           --> 0 to 63 bits
    
    key_permutation_matrix = [57, 49, 41, 33, 25, 17, 9,
                               1, 58, 50, 42, 34, 26, 18,
                              10, 2, 59, 51, 43, 35, 27,
                              19, 11, 3, 60, 52, 44, 36,
                              63, 55, 47, 39, 31, 23, 15,
                               7, 62, 54, 46, 38, 30, 22,
                              14, 6, 61, 53, 45, 37, 29,
                              21, 13, 5, 28, 20, 12, 4 ]

    permuted_matrix = [0 for i in range(56)]
    
    for i in range(0, len(key_permutation_matrix)):
        index              = key_permutation_matrix[i] - 1
        permuted_matrix[i] = elements[index]
        
    return permuted_matrix
```

## left, right split

```python
def left_right_split(matrix):
    left_str  = matrix[:28]
    right_str = matrix[28:]
    
    return (left_str, right_str)
```

```python
def left_shift(left_str, right_str):  ## left shifting
    list1 = []
    list1.append((left_str, right_str))  ## C0 and D0
    
    for i in range(1,17): 
        if (i==1 or i==2 or i==9 or i==16): ## 1-bit shift (C1 and D1 ; C2 and D2 ; C9 and D9 ; C16 and D16)
            right_str = right_str[1:] + list(right_str[-1])
            left_str = left_str[1:] + list(left_str[-1])
            list1.append((left_str, right_str))
            
        else: ## 2-bit shift
            right_str = right_str[2:] + list(right_str[-1]) + list(right_str[-1])
            left_str = left_str[2:] + list(left_str[-1]) + list(left_str[-1])
            list1.append((left_str, right_str))
    
    return list1
```

```python
key_combination = left_shift(left_str, right_str)
```

```python
cnt=0
for i in key_combination:
    print("C{} --> ".format(cnt), end='')
    display_7(i[0])
    
    print("\n")
    print("D{} --> ".format(cnt), end='')
    display_7(i[1])
    print("\n\n")
    cnt+=1
```

## Expansion D-box Table

```python
def expansion_permutation(left_str):
    exp_d = [32, 1 , 2 , 3 , 4 , 5 , 4 , 5,
             6 , 7 , 8 , 9 , 8 , 9 , 10, 11,
             12, 13, 12, 13, 14, 15, 16, 17,
             16, 17, 18, 19, 20, 21, 20, 21,
             22, 23, 24, 25, 24, 25, 26, 27,
             28, 29, 28, 29, 30, 31, 32, 1 ]
    
    expansion_matrix = [0 for i in range(64)]
    for i in range(0, len(exp_d)):
        index = exp_d[i] 
        expansion_matrix[i] = left_str[index]
        
    return expansion_matrix
```

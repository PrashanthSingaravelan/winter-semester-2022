```python
import numpy as np
import pandas as pd

import scipy.stats as stats
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

<body style="background-color:FFFF33;">
<h1 style="text-align:left; font-family:serif; font-size:200%">Importing the Data-Set</h1>

```python
df = pd.read_excel('Procrastination self Assessment form.xls')
df.head()
```

```python
print("Number of response collected : ",len(df))
```

<body style="background-color:FFFF33;">
 <h1 style="text-align:center; font-family:serif; font-size:300%">Dealing with objective-1</h1>

```python
question_1 = df.iloc[:,2:6]
question_1.head()
```

```python
question_1['index'] = question_1.index
question_1.head()
```

```python
print("Answers Options :  ",len(question_1.columns))
```

<body style="background-color:FFFF33;">
<h1 style="text-align:left; font-family:serif; font-size:200%">Dividing into Activists and Non Activists</h1>

```python
activists = question_1[(question_1["I often watch movies night"] > 4) |
                                   (question_1["If it is time to turn off the lights at night, I overthink what I read/watch before sleeping"] > 3)]

print(len(activists))
activists.head()
```

```python
non_activists = question_1[question_1.index.isin(activists.index) == False]

print(len(non_activists))
non_activists.head()
```

```python
activists.drop('index', axis='columns', inplace=True)
activists.loc[:,'Row_Total']        = activists.sum(numeric_only=True, axis=1)

non_activists.drop('index', axis='columns', inplace=True)
non_activists.loc[:,'Row_Total'] = non_activists.sum(numeric_only=True, axis=1)
```

<body style="background-color:FFFF33;">
    <h1 style="text-align:left; font-family:serif; font-size:200%">Perform F-Test</h1>

```python
variable_1 = activists.sample(n=30)['Row_Total']
variable_2 = non_activists.sample(n=30)['Row_Total']
```

```python
def f_test(x, y):
    x = np.array(x)
    y = np.array(y)
    f = np.var(x, ddof=1)/np.var(y, ddof=1) #calculate F test statistic 
    dfn = x.size-1                    #define degrees of freedom numerator 
    dfd = y.size-1 #define degrees of freedom denominator 
    p = 1-stats.f.cdf(f, dfn, dfd) #find p-value of F test statistic 
    return f, p
```

```python
f_value, p_value = f_test(variable_1, variable_2)

print(f_value)
print(p_value)
```

```python
if p_value < 0.05:   
    print("Rejecting Null Hypothesis (i.e) Accepting Alternate Hypothesis")
else:
    print("Accepting Null Hypothesis (i.e) Rejecting Alternate Hypothesis")
```

```python
question_2 = df.iloc[:,7:12]

question_2.head()
```

<body style="background-color:FFFF33;">
<h1 style="text-align:left; font-family:serif; font-size:200%">Dividing into People Around Us are Procastinating / Not</h1>

```python
ar_procastinated = question_2[
                                    (question_2['I get inspired by my friend circle work punctuality'] > 4) | 
                                    (question_2['I postpone my work due to sudden/unexpected plans with my circle'] > 3) |
                                    (question_2['I get demotivated on comparing myself with others and postpone my works'] > 4)
                            ]

print(len(ar_procastinated))
ar_procastinated.head()
```

```python
non_ar_procastinated = question_2[question_2.index.isin(ar_procastinated.index) == False]

print(len(non_ar_procastinated))
non_ar_procastinated.head()
```

```python
ar_procastinated.drop('index', axis='columns', inplace=True)
ar_procastinated.loc[:,'Row_Total']        = ar_procastinated.sum(numeric_only=True, axis=1)

non_ar_procastinated.drop('index', axis='columns', inplace=True)
non_ar_procastinated.loc[:,'Row_Total'] = non_ar_procastinated.sum(numeric_only=True, axis=1)
```

```python
variable_1 = ar_procastinated.sample(n=30)['Row_Total']
variable_2 = non_ar_procastinated.sample(n=30)['Row_Total']
```

<body style="background-color:FFFF33;">
    <h1 style="text-align:left; font-family:serif; font-size:200%">Perform F-Test</h1>

```python
f_value, p_value = f_test(variable_1, variable_2)

print("f-value : {}".format(f_value),end = ' ')
print("and p-value : ",p_value)
```

```python
if p_value < 0.05:   
    print("Rejecting Null Hypothesis (i.e) Accepting Alternate Hypothesis")
else:
    print("Accepting Null Hypothesis (i.e) Rejecting Alternate Hypothesis")
```

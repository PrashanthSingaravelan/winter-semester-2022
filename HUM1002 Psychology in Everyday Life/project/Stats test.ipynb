```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_excel('Procrastination self Assessment form.xls')
df.head()
```

```python
print("Number of response collected : ",len(df))
```

## Dealing with question_1

```python
question_1 = df.iloc[:,2:6]
question_1.head()
```

```python
print("Answers Options :  ",len(question_1.columns))
```

```python
question_1.loc['Column_Total'] = question_1.sum(numeric_only=True, axis=0)
question_1.loc[:,'Row_Total']    = question_1.sum(numeric_only=True, axis=1)

question_1.tail()
```
